```python
### Load filtered BNDs ###

##
BND_path = '/Users/ryanyutian/Desktop/Manuscript/filtered_sv/BND/'

BND_raw_df_names = []

os.chdir(BND_path)
temp_files = sorted([i for i in os.listdir(BND_path) if 'DS' not in i])

for file_name in temp_files:
    
    globals()[file_name[:-4]] = pd.read_csv(file_name)
    BND_raw_df_names.append(file_name[:-4])
```

```python
### Load filtered somatic large inversions ###

##
somatic_large_INV_path = '/Users/ryanyutian/Desktop/Manuscript/filtered_sv/INV/'

somatic_large_INV_filtered_df_names = []

os.chdir(somatic_large_INV_path)
temp_files = sorted([i for i in os.listdir(somatic_large_INV_path) if 'DS' not in i])

for file_name in temp_files:
    
    globals()[file_name[:-4]] = pd.read_csv(file_name)
    somatic_large_INV_filtered_df_names.append(file_name[:-4])
```

## Preprocess inversion dfs

```python
processed_inv_df_names = []

for df_name in somatic_large_INV_filtered_df_names:
    
    output_data = []
    
    temp_df = globals()[df_name]
    
    for _, row in temp_df.iterrows():
        
        output_data.append([row['CHROM'], row['POS'], row['POS'], \
                            row['CHROM'], row['END'], row['END']])
        
    output_df = pd.DataFrame(output_data, columns=['CHROM1', 'POS1', 'END1', 'CHROM2', 'POS2', 'END2'])
    
    globals()[df_name[:-4]+'_INV_processed_CIRCOS'] = output_df.copy()
    
    processed_inv_df_names.append(df_name[:-4]+'_INV_processed_CIRCOS')
```

# Save individual samples

```python
# Inversion

for df_name in processed_inv_df_names:
    
    temp_df = globals()[df_name]
    
    temp_df.to_csv(('/Users/ryanyutian/Desktop/Manuscript/CIRCOS/individual_data/INV/' + \
                    df_name[:-17] + '.csv'), index=False, sep=',')
```

```python
# Translocation

for df_name in BND_raw_df_names:
    
    temp_df = globals()[df_name]
    
    temp_tl_df = temp_df[temp_df['CHROM1'] != temp_df['CHROM2']]

    selected_columns = ['CHROM1', 'POS1', 'CHROM2', 'POS2']
    save_df = temp_tl_df[selected_columns].copy()

    # Add 'END1' and 'END2' columns
    save_df['END1'] = save_df['POS1']
    save_df['END2'] = save_df['POS2']

    save_df = save_df[['CHROM1', 'POS1', 'END1', 'CHROM2', 'POS2', 'END2']]
    
    save_df.to_csv(('/Users/ryanyutian/Desktop/Manuscript/CIRCOS/individual_data/BND/' + \
                     df_name + '.csv'), index=False, sep=',')
```
