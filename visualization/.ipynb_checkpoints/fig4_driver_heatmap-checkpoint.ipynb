# 2. Compile df with all SV types for plotting

## Nomenclature:

### 1=DEL, 2=DUP, 3=INV, 4=DEL+DUP, 5=DEL+INV, 6=DUP+INV, 7=ALL

```python
glioma_driver_all_count_df = \
pd.DataFrame(0, index=glioma_driver_DEL_count_df.index, columns=glioma_driver_DEL_count_df.columns)

glioma_driver_all_count_df
```

```python
def determine_category(gene, sample_id, DEL_df, DUP_df, INV_df):
    """
    
    Determines the category of structural variants present for a given gene and sample ID.

    Parameters:
    - gene (str): The gene identifier.
    - sample_id (str): The sample identifier.
    - DEL_df (DataFrame): A pandas DataFrame containing deletion data.
    - DUP_df (DataFrame): A pandas DataFrame containing duplication data.
    - INV_df (DataFrame): A pandas DataFrame containing inversion data.

    Returns:
    int: Category number indicating the type of structural variants detected:
        - 1 for DEL only
        - 2 for DUP only
        - 3 for INV only
        - 4 for DEL and DUP
        - 5 for DEL and INV
        - 6 for DUP and INV
        - 7 for all types (DEL, DUP, INV)
        - 0 if no structural variants are detected.
    
    """

    if DEL_df.loc[gene, sample_id] > 0:
        
        if DUP_df.loc[gene, sample_id] > 0:
                    
            if INV_df.loc[gene, sample_id] > 0:
                
                # ALL SV TYPE (7)
                return 7
                
            else:
                
                # DEL + DUP (4)
                return 4

        elif INV_df.loc[gene, sample_id] > 0:
            
            # DEL + INV (5)
            return 5
            
        else:
            
            # DEL ONLY (1)
            return 1
            
    elif DUP_df.loc[gene, sample_id] > 0:
        
        if INV_df.loc[gene, sample_id] > 0:
        
            # DUP + INV (6)
            return 6
            
        else:
            
            # DUP ONLY (2)
            return 2
            
    elif INV_df.loc[gene, sample_id] > 0:
            
            # INV ONLY (3)
            return 3
        
    else:
        
        return 0
```

```python
for gene, row in glioma_driver_all_count_df.iterrows():
    
    for sample_id in glioma_driver_all_count_df.columns:
        
        glioma_driver_all_count_df.loc[gene, sample_id] = \
        determine_category(gene, sample_id, glioma_driver_DEL_count_df, \
                           glioma_driver_DUP_count_df, glioma_driver_INV_count_df)
```

```python
def extract_chrom_arm_name(cytoband):

    match = re.search(r'^(.*?[pq])', cytoband)

    if match:
        
        return match.group(1)
    
    else:

        return 'error'
```

```python
chrom_arm_dict = {}

for gene in gene_cytoband_dict:
    
    temp_chrom_arm = extract_chrom_arm_name(gene_cytoband_dict[gene][0])
    
    if temp_chrom_arm in chrom_arm_dict:
        
        chrom_arm_dict[temp_chrom_arm].append(gene)
    
    else:
        
        chrom_arm_dict[temp_chrom_arm] = [gene]
```

# 4. Plotting SV heatmap

```python
# Function to add half colored rectangles in the heatmap

def add_half_colored_rectangle(ax, bbox, color1, color2):
    
    rect1 = plt.Rectangle((bbox[0], bbox[1]), 0.5, 1, facecolor=color1, edgecolor='gray', linewidth=.5)
    rect2 = plt.Rectangle((bbox[0] + 0.5, bbox[1]), 0.5, 1, facecolor=color2, edgecolor='gray', linewidth=.5)
    ax.add_patch(rect1)
    ax.add_patch(rect2)
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(20, 58), dpi=300)

mpl.rcParams['font.family'] = 'sans-serif'

# Define the color map for the heatmap
cmap = {0: 'white', 1: 'blue', 2: 'red', 3: 'green', \
        4: 'white', 5: 'white', 6: 'white', 7: 'purple'}

colors = [cmap[i] for i in range(0, 8)]

heatmap_cmap = mpl.colors.ListedColormap(colors)


# Create the heatmap
sns.heatmap(glioma_driver_toplot_burden_order_df, cmap=heatmap_cmap, ax=ax, linewidths=.5, linecolor='gray', 
            cbar=False, square=True, vmin=0, vmax=7)


# Iterate through heatmap data and add colored rectangles
for y in range(len(glioma_driver_toplot_burden_order_df)):
    for x in range(len(glioma_driver_toplot_burden_order_df.columns)):
        value = glioma_driver_toplot_burden_order_df.iloc[y, x]
        if value == 4:
            add_half_colored_rectangle(ax, (x, y), 'blue', 'red')
        elif value == 5:
            add_half_colored_rectangle(ax, (x, y), 'blue', 'green')
        elif value == 6:
            add_half_colored_rectangle(ax, (x, y), 'red', 'green')

# Move the y-axis labels to the left and x-axis labels to the top
ax.yaxis.tick_left()
ax.xaxis.tick_top()

# Set the tick label color and size
ax.tick_params(axis='both', colors='black', labelsize=18)
for label in ax.get_yticklabels():
    label.set_fontstyle('italic')

# Remove the ticks while keeping the labels
ax.yaxis.set_tick_params(width=0)
ax.xaxis.set_tick_params(width=0)

# Rotate x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation='vertical')


# Draw a border around the heatmap
heatmap_outline = mpatches.Rectangle((-0.01, -0.02), len(glioma_driver_toplot_burden_order_df.columns), \
                                     len(glioma_driver_toplot_burden_order_df.index), fill=None, \
                                     edgecolor='gray', linewidth=0.5)

ax.add_patch(heatmap_outline)

plt.tight_layout()

fig.savefig('/Users/ryanyutian/Desktop/Manuscript/driver_heatmap/driver_heatmap_burden_order.png', dpi=300, bbox_inches='tight')
```
