```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
burden_df = pd.read_csv('/Users/ryanyutian/Desktop/Manuscript/burden/burden_w_new_ID.csv')

burden_df
```

```python
burden_df['patient_id'].unique().tolist()
```

```python
DEL_burden_dict = {}
DUP_burden_dict = {}
INV_burden_dict = {}

for p_id in burden_df['patient_id'].unique().tolist():
    
    if p_id != 'X':

        DEL_burden_dict[p_id] = []
        DUP_burden_dict[p_id] = []
        INV_burden_dict[p_id] = []
        
        for i in range(len(['Primary', 'Recurrence', 'Second Recurrence'])):

            DEL_burden_dict[p_id].append('N/A')
            DUP_burden_dict[p_id].append('N/A')
            INV_burden_dict[p_id].append('N/A')

###

X_counter = 1

for _, row in burden_df.iterrows():
    
    
    if row['patient_id'] != 'X':
        
        DEL_burden_dict[row['patient_id']][row['stage_rank']] = row['deletion']
        DUP_burden_dict[row['patient_id']][row['stage_rank']] = row['duplication']
        INV_burden_dict[row['patient_id']][row['stage_rank']] = row['inversion']
        
    else:
        
        DEL_burden_dict['X' + str(X_counter)] = ['N/A', 'N/A', 'N/A']
        DUP_burden_dict['X' + str(X_counter)] = ['N/A', 'N/A', 'N/A']
        INV_burden_dict['X' + str(X_counter)] = ['N/A', 'N/A', 'N/A']

        DEL_burden_dict['X' + str(X_counter)][row['stage_rank']] = row['deletion']
        DUP_burden_dict['X' + str(X_counter)][row['stage_rank']] = row['duplication']
        INV_burden_dict['X' + str(X_counter)][row['stage_rank']] = row['inversion']
        
        X_counter += 1
```

```python
INV_burden_dict
```

```python
total_burden_dict = {}

for p_id in DEL_burden_dict:
    
    total_burden_dict[p_id] = []
    
    for i in range(len(DEL_burden_dict[p_id])):
        
        if DEL_burden_dict[p_id][i] != 'N/A':
            
            total_burden_dict[p_id]\
            .append(DEL_burden_dict[p_id][i] + DUP_burden_dict[p_id][i] + INV_burden_dict[p_id][i])
            
        else:
            
            total_burden_dict[p_id].append('N/A')
```

```python
total_burden_dict
```

```python
# Setup plot appearance
sns.set(style='white', rc={"axes.facecolor": "white", "grid.color": "black"})

labels = ['Primary', 'Recurrence', 'Second\nRecurrence']
label_indices = {label: i for i, label in enumerate(labels)}

fig, ax = plt.subplots(figsize=(10, 5), dpi=300)

for key, values in total_burden_dict.items():
    x = []
    y = []
    for i, value in enumerate(values):
        if value != 'N/A':
            x.append(label_indices[labels[i]])
            y.append(value)
    ax.scatter(x, y, label=key, s=60, color='purple')
    if len(x) > 1:
        ax.plot(x, y, linestyle='--', color='gray')

# Set the x-ticks to be categorical
ax.set_xticks(list(label_indices.values()))
ax.set_xticklabels(labels, fontsize=15, fontweight='bold')

ax.tick_params(axis='y', labelsize=19)
ax.tick_params(axis='both', which='both', length=5, width=1, direction='out', colors='black')

# y-axis tick mark settings
for tick in ax.yaxis.get_major_ticks():
    tick.tick1line.set_visible(True)
    tick.tick1line.set_linewidth(1)
    tick.tick1line.set_color('black')

# x-axis tick mark settings
for tick in ax.xaxis.get_major_ticks():
    tick.tick1line.set_visible(True)
    tick.tick1line.set_linewidth(1)
    tick.tick1line.set_color('black')

ax.set_xlabel('Tumour Stage', weight='bold', labelpad=10, fontsize=10)
ax.tick_params(axis='x', pad=10)  
ax.set_ylim(0, 801)
ax.set_yticks(np.arange(0, 801, 200))

plt.ylabel('SV Burden\nDeletions (Mb)', weight='bold', labelpad=50, fontsize=10, rotation=0, ha='center', color='black')
plt.tight_layout()

fig.savefig('/Users/ryanyutian/Desktop/Manuscript/burden/all_pri_rec.png', dpi=300, bbox_inches='tight')
```
